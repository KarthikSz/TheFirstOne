```python
from nltk.corpus import stopwords
def text_processor(message):
    nopunc=[c for c in message if c not in string.punctuation]
    nopunc="".join(nopunc)
    clean_mess=[word for word in nopunc.split() if word not in stopwords.words('english') ]
    return clean_mess
```

Below I'm gonna try to seperate the training for gossip and political to check the accuracy

```python
df_gossip_overall=pd.concat([df_gr,df_gf])
```

```python
df_political_overall=pd.concat([df_pr,df_pf])
```

Creating 2 seperate pipelines for both:

```python
pipe_lino_2=Pipeline([
                    ('Step 1',bow_transformer),
                    ('Step 2',tf_idf_transformer),
                    ('Step 3',RandomForestClassifier())
])
```

```python
pipe_lino_3=Pipeline([
                    ('Step 1',bow_transformer),
                    ('Step 2',tf_idf_transformer),
                    ('Step 3',RandomForestClassifier())
])
```

```python
from sklearn.model_selection import train_test_split
```

```python
gtit_train,gtit_test,glab_train,glab_test=train_test_split(df_gossip_overall['title'],df_gossip_overall['label'])
```

```python
ptit_train,ptit_test,plab_train,plab_test=train_test_split(df_political_overall['title'],df_political_overall['label'])
```

```python
pipe_lino_2.fit(gtit_train,glab_train)
```

```python
pipe_lino_3.fit(ptit_train,plab_train)
```

```python
predictions_2=pipe_lino_2.predict(gtit_test)
```

```python
predictions_3=pipe_lino_3.predict(ptit_test)
```

```python
print(classification_report(predictions_2,glab_test))
```

```python
print(classification_report(predictions_3,plab_test))
```

Hence, by seperating gossip and other columns we can see a change in the accuracy
